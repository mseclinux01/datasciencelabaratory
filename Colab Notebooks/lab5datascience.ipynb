#(a) Univariate analysis:
    The term univariate analysis refers to the analysis of one variable. You can remember this because the prefix “uni” means “one.”
**There are three common ways to perform univariate analysis on one variable:**
 - Summary statistics – Measures the center and spread of values.
 -Frequency table – Describes how often different values occur.
 - Charts – Used to visualize the distribution of values.

#(b). Bivariate analysis: Linear and logistic regression modeling.

```python
df.corr()
```

```python
#Simple Linear Regression
import statsmodels.api as sm

#define response variable
y = df['Pregnancies']

#define explanatory variable
x = df[['Glucose']]

#add constant to predictor variables
x = sm.add_constant(x)

#fit linear regression model
model = sm.OLS(y, x).fit()

#view model summary
print(model.summary())
```

```python
#Simple Linear Regression
import statsmodels.api as sm
#Importing CSV Files From local computer
from google.colab import files #importing libraries
uploaded = files.upload()#uploading from local pc

df = pd.read_csv("diabetes.csv")

#define response variable
y = df[['Pregnancies']]

#define explanatory variable
x = df[['Glucose']]

#add constant to predictor variables
x = sm.add_constant(x)

#fit linear regression model
model = sm.OLS(y, x).fit()

#view model summary
print(model.summary())
```
